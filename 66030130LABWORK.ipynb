### แบบฝึกหัดที่ 2: Array 1 มิติ

1. จงเขียนโปรแกรมสร้าง Array เก็บคะแนนสอบของนักเรียน 5 คน และ:
   - หาค่าเฉลี่ย
   - หาคะแนนสูงสุดและต่ำสุด
   - แสดงจำนวนคนที่ได้คะแนนมากกว่าค่าเฉลี่ย

```python
student_score = []
for i in range(5):
    score = float(input(f"กรุณาป้อนคะแนนของนักเรียนคนที่ {i+1} : "))
    student_score.append(score)
average = sum(student_score) / len(student_score)
score_max = max(student_score)
score_min = min(student_score)
above_average_count = sum(student_score > average for student_score in student_score)

print(f"นักเรียน 5 คนมีคะแนนอยุ่ดังนี้ : {student_score}")
print(f"คะแนนสูงสุด : {score_max}")
print(f"คะแนนต่ำสุด : {score_min}")
print(f"ค่าเฉลี่ย : {average}")
print(f"จำนวนคนที่ได้คะแนนมากกว่าค่าเฉลี่ย : {above_average_count} คน")
```

2. สร้างโปรแกรมจัดการรายชื่อสินค้าในร้านค้า โดยมีฟังก์ชัน:
   - เพิ่มสินค้าใหม่
   - ลบสินค้าที่หมด
   - ค้นหาสินค้า
   - แสดงรายการสินค้าทั้งหมด

```python
stock = []

def add_product():
    while True:
        new_product = input("เพิ่มสินค้าชิ้นใหม่ :")
        stock.append(new_product)
        print(f"เพิ่มสินค้า {new_product} เรียบร้อยแล้ว")
        continue_adding = input("ต้องการเพิ่มสินค้าชิ้นอื่นหรือไม่? (yes/no): ").lower()
        if continue_adding != "yes":
            break
    
def remove_product():
    while True:
        remove_product = input("เลือกสินค้าที่ต้องการจะลบ :")
        if remove_product in stock:
            stock.remove(remove_product)
            print(f"ลบ {remove_product} ออกจากสต็อกเรียบร้อยแล้ว")
        else:
            print(f"ไม่พบ {remove_product} ในสต็อก")
            
        continue_removing = input("ต้องการลบสินค้าชิ้นอื่นหรือไม่? (yes/no): ").lower()
        if continue_removing != "yes":
            break

def search_product():
    search_product = input("ชื่่อสินค้าที่ต้องการค้นหา :")
    if search_product in stock:
        print(f"เจอ {search_product} ในสต็อกแล้ว")
    else:
        print(f"ไม่พบ {search_product} ในสต็อก")

def list_product():
    print("สินค้าในสต๊อกปัจจุบันมีดังนี้ :")
    for index,product in enumerate(stock, start=1):
        print(f"{index}.{product}")

def menu():
    while True:
        print("พิมพ์ 1 เพื่อเพิ่มสินค้า")
        print("พิมพ์ 2 เพื่อลบสินค้า")
        print("พิมพ์ 3 เพื่อค้นหาสินค้า")
        print("พิมพ์ 4 เพื่อแสดงรายการสินค้าทั้งหมด")
        print("พิมพ์ 5 เพื่อออกจากโปรแกรม")
    
        choice = input("กรุณาเลือกทำรายการ :")
        
        if choice == "1" :
            add_product()
        elif choice == "2" :
            remove_product()
        elif choice == "3" :
            search_product()
        elif choice == "4" :
            list_product()
        elif choice == "5" :
            yes_or_no = input("ต้องการออกจากโปรแกรมใช่หรือไม่ (yes/no) :")
            if yes_or_no == "yes" :
                print("ออกจากโปรแกรมเรียบร้อยแล้ว")
                break
            else:
                continue
        else:
            print("กรุณาเลือกทำรายการใหม่อีกครั้ง")

if __name__ == "__main__":
    menu()
```

### แบบฝึกหัดที่ 2: Array 2 มิติ
1. จงเขียนโปรแกรมจัดการคะแนนสอบ:
   - เก็บคะแนนสอบ 3 วิชาของนักเรียน 4 คน
   - คำนวณคะแนนเฉลี่ยของแต่ละวิชา
   - หานักเรียนที่ได้คะแนนรวมสูงสุด
   - แสดงจำนวนนักเรียนที่สอบผ่านแต่ละวิชา (คะแนนผ่าน ≥ 50)

```python
scores = [
    [65, 70, 80],  
    [45, 55, 60],  
    [75, 85, 90],
    [50, 60, 70]   
]

num_students = len(scores)
num_subjects = len(scores[0])
subject_averages = [0] * num_subjects

for subject in range(num_subjects):
    total = sum(scores[student][subject] for student in range(num_students))
    subject_averages[subject] = total / num_students

student_totals = [sum(student_scores) for student_scores in scores]
max_score = max(student_totals)
top_student = student_totals.index(max_score) + 1

pass_counts = [0] * num_subjects

for subject in range(num_subjects):
    pass_counts[subject] = sum(1 for student in range(num_students) if scores[student][subject] >= 50)

print("คะแนนเฉลี่ยของแต่ละวิชา:")
for i, avg in enumerate(subject_averages, start=1):
    print(f"วิชา {i}: {avg:.2f}")

print("\nนักเรียนที่ได้คะแนนรวมสูงสุด:")
print(f"นักเรียนหมายเลข {top_student} ได้คะแนนรวม {max_score}")

print("\nจำนวนนักเรียนที่สอบผ่านในแต่ละวิชา:")
for i, count in enumerate(pass_counts, start=1):
    print(f"วิชา {i}: {count} คน")
```

2. สร้างโปรแกรมจัดการที่นั่งในโรงภาพยนตร์:
   - แสดงแผนผังที่นั่ง (ว่าง/จอง)
   - จองที่นั่ง
   - ยกเลิกการจอง
   - แสดงจำนวนที่นั่งว่างทั้งหมด

```python
# สร้างแผนผังที่นั่ง (10 แถว, แถวละ 10 ที่นั่ง)
rows = 10
cols = 10
seats = [["ว่าง" for _ in range(cols)] for _ in range(rows)]

def display_seats():
    """แสดงแผนผังที่นั่ง"""
    print("\nแผนผังที่นั่ง:")
    print("   " + " ".join([f"{i+1:2}" for i in range(cols)]))  # แสดงหมายเลขคอลัมน์
    for i, row in enumerate(seats):
        print(f"{i+1:2} " + " ".join(row))  # แสดงหมายเลขแถวและสถานะที่นั่ง
    print()

def book_seat():
    """จองที่นั่ง"""
    try:
        row = int(input("กรุณากรอกหมายเลขแถว (1-10): ")) - 1
        col = int(input("กรุณากรอกหมายเลขที่นั่งในแถว (1-10): ")) - 1
        if seats[row][col] == "ว่าง":
            seats[row][col] = "จอง"
            print("จองที่นั่งเรียบร้อยแล้ว\n")
        else:
            print("ที่นั่งนี้ถูกจองไปแล้ว\n")
    except (IndexError, ValueError):
        print("กรุณากรอกหมายเลขที่นั่งให้ถูกต้อง (1-10)\n")

def cancel_seat():
    """ยกเลิกการจองที่นั่ง"""
    try:
        row = int(input("กรุณากรอกหมายเลขแถว (1-10): ")) - 1
        col = int(input("กรุณากรอกหมายเลขที่นั่งในแถว (1-10): ")) - 1
        if seats[row][col] == "จอง":
            seats[row][col] = "ว่าง"
            print("ยกเลิกการจองเรียบร้อยแล้ว\n")
        else:
            print("ที่นั่งนี้ยังไม่ได้ถูกจอง\n")
    except (IndexError, ValueError):
        print("กรุณากรอกหมายเลขที่นั่งให้ถูกต้อง (1-10)\n")

def count_available_seats():
    """แสดงจำนวนที่นั่งว่างทั้งหมด"""
    available_seats = sum(row.count("ว่าง") for row in seats)
    print(f"\nจำนวนที่นั่งว่างทั้งหมด: {available_seats} ที่\n")

# เมนูหลัก
while True:
    print("โปรแกรมจัดการที่นั่งในโรงภาพยนตร์")
    print("1. แสดงแผนผังที่นั่ง")
    print("2. จองที่นั่ง")
    print("3. ยกเลิกการจอง")
    print("4. แสดงจำนวนที่นั่งว่างทั้งหมด")
    print("5. ออกจากโปรแกรม")

    choice = input("กรุณาเลือกเมนู: ")
    if choice == '1':
        display_seats()
    elif choice == '2':
        book_seat()
    elif choice == '3':
        cancel_seat()
    elif choice == '4':
        count_available_seats()
    elif choice == '5':
        print("ออกจากโปรแกรม...")
        break
    else:
        print("กรุณาเลือกเมนูให้ถูกต้อง (1-5)\n")
```

### แบบฝึกหัดที่ 3: Array 3 มิติ

1. จงเขียนโปรแกรมจัดการข้อมูลอุณหภูมิ:
   - เก็บอุณหภูมิของ 3 เมือง
   - วัด 4 ครั้งต่อวัน
   - เป็นเวลา 7 วัน
   - หาค่าเฉลี่ยอุณหภูมิของแต่ละเมือง
   - หาวันและเวลาที่อุณหภูมิสูงที่สุดของแต่ละเมือง

```python
# สร้าง Array 3 มิติ [เมือง][วัน][เวลา]
cities = 3
days = 7
times = 4
temperatures = [[[0 for _ in range(times)] for _ in range(days)] for _ in range(cities)]

def input_temperatures():
    """ให้ผู้ใช้ป้อนข้อมูลอุณหภูมิ"""
    for city in range(cities):
        print(f"\nกรุณาป้อนอุณหภูมิของเมืองที่ {city + 1}:")
        for day in range(days):
            for time in range(times):
                temperatures[city][day][time] = float(input(f"  วันที่ {day + 1}, เวลา {time + 1}: "))

def display_temperatures():
    """แสดงอุณหภูมิทั้งหมด"""
    for city in range(cities):
        print(f"\nเมืองที่ {city + 1}:")
        for day in range(days):
            print(f"  วันที่ {day + 1}: ", end="")
            for time in range(times):
                print(f"{temperatures[city][day][time]:.1f}°C", end=" ")
            print()

def calculate_average_temperatures():
    """คำนวณค่าเฉลี่ยอุณหภูมิของแต่ละเมือง"""
    averages = []
    for city in range(cities):
        total = sum(temperatures[city][day][time] for day in range(days) for time in range(times))
        avg = total / (days * times)
        averages.append(avg)
        print(f"ค่าเฉลี่ยอุณหภูมิของเมืองที่ {city + 1}: {avg:.2f}°C")
    return averages

def find_highest_temperatures():
    """หาวันและเวลาที่อุณหภูมิสูงที่สุดของแต่ละเมือง"""
    for city in range(cities):
        max_temp = float('-inf')
        max_day, max_time = -1, -1
        for day in range(days):
            for time in range(times):
                if temperatures[city][day][time] > max_temp:
                    max_temp = temperatures[city][day][time]
                    max_day, max_time = day, time
        print(f"เมืองที่ {city + 1}: อุณหภูมิสูงสุด {max_temp:.1f}°C ในวันที่ {max_day + 1} เวลา {max_time + 1}")

# เมนูหลัก
while True:
    print("\nโปรแกรมจัดการข้อมูลอุณหภูมิ")
    print("1. ป้อนข้อมูลอุณหภูมิ")
    print("2. แสดงอุณหภูมิทั้งหมด")
    print("3. คำนวณค่าเฉลี่ยอุณหภูมิของแต่ละเมือง")
    print("4. หาวันและเวลาที่อุณหภูมิสูงที่สุดของแต่ละเมือง")
    print("5. ออกจากโปรแกรม")

    choice = input("กรุณาเลือกเมนู: ")
    if choice == '1':
        input_temperatures()
    elif choice == '2':
        display_temperatures()
    elif choice == '3':
        calculate_average_temperatures()
    elif choice == '4':
        find_highest_temperatures()
    elif choice == '5':
        print("ออกจากโปรแกรม...")
        break
    else:
        print("กรุณาเลือกเมนูให้ถูกต้อง (1-5)\n")
```

2. เขียนโปรแกรมวิเคราะห์ผลการเรียน:
   - เก็บคะแนนของนักเรียน 3 ห้อง
   - ห้องละ 5 คน
   - สอบ 4 วิชา
   - หาห้องที่มีคะแนนเฉลี่ยสูงสุด
   - หานักเรียนที่ได้คะแนนรวมสูงสุดในแต่ละห้อง
   - แสดงจำนวนนักเรียนที่สอบผ่านทุกวิชาในแต่ละห้อง

```python
# กำหนดขนาดของข้อมูล
rooms = 3
students_per_room = 5
subjects = 4

# สร้าง array ว่างสำหรับเก็บคะแนน
scores = [[[0 for _ in range(subjects)] for _ in range(students_per_room)] for _ in range(rooms)]

def input_scores():
    """ให้ผู้ใช้ป้อนคะแนนของนักเรียนในแต่ละห้อง"""
    for room in range(rooms):
        print(f"\nกรุณาป้อนคะแนนของห้องที่ {room + 1}:")
        for student in range(students_per_room):
            print(f"  นักเรียนที่ {student + 1}:")
            for subject in range(subjects):
                scores[room][student][subject] = float(input(f"    วิชาที่ {subject + 1}: "))

def display_scores():
    """แสดงคะแนนทั้งหมด"""
    for room in range(rooms):
        print(f"\nคะแนนของห้องที่ {room + 1}:")
        for student in range(students_per_room):
            print(f"  นักเรียนที่ {student + 1}: {scores[room][student]}")

def calculate_average_per_room():
    """คำนวณค่าเฉลี่ยคะแนนของแต่ละห้อง"""
    room_averages = []
    for room in range(rooms):
        total_score = sum(scores[room][student][subject] for student in range(students_per_room) for subject in range(subjects))
        avg = total_score / (students_per_room * subjects)
        room_averages.append(avg)
        print(f"ค่าเฉลี่ยคะแนนของห้องที่ {room + 1}: {avg:.2f}")
    return room_averages

def find_highest_average_room():
    """หาห้องที่มีคะแนนเฉลี่ยสูงสุด"""
    room_averages = calculate_average_per_room()
    max_avg = max(room_averages)
    best_room = room_averages.index(max_avg) + 1
    print(f"\nห้องที่มีคะแนนเฉลี่ยสูงสุดคือห้องที่ {best_room} ด้วยคะแนนเฉลี่ย {max_avg:.2f}")

def find_highest_scorer_in_each_room():
    """หานักเรียนที่ได้คะแนนรวมสูงสุดในแต่ละห้อง"""
    for room in range(rooms):
        max_score = -1
        best_student = -1
        for student in range(students_per_room):
            total_score = sum(scores[room][student])
            if total_score > max_score:
                max_score = total_score
                best_student = student + 1
        print(f"ในห้องที่ {room + 1}: นักเรียนที่ได้คะแนนรวมสูงสุดคือคนที่ {best_student} ด้วยคะแนน {max_score}")

def count_students_passing_all_subjects():
    """แสดงจำนวนนักเรียนที่สอบผ่านทุกวิชาในแต่ละห้อง"""
    for room in range(rooms):
        passing_students = 0
        for student in range(students_per_room):
            if all(score >= 50 for score in scores[room][student]):
                passing_students += 1
        print(f"ในห้องที่ {room + 1}: มีนักเรียน {passing_students} คนที่สอบผ่านทุกวิชา")

# เมนูหลัก
while True:
    print("\nโปรแกรมวิเคราะห์ผลการเรียน")
    print("1. ป้อนคะแนน")
    print("2. แสดงคะแนนทั้งหมด")
    print("3. คำนวณค่าเฉลี่ยคะแนนของแต่ละห้อง")
    print("4. หาห้องที่มีคะแนนเฉลี่ยสูงสุด")
    print("5. หานักเรียนที่ได้คะแนนรวมสูงสุดในแต่ละห้อง")
    print("6. แสดงจำนวนนักเรียนที่สอบผ่านทุกวิชาในแต่ละห้อง")
    print("7. ออกจากโปรแกรม")

    choice = input("กรุณาเลือกเมนู: ")

    if choice == '1':
        input_scores()
    elif choice == '2':
        display_scores()
    elif choice == '3':
        calculate_average_per_room()
    elif choice == '4':
        find_highest_average_room()
    elif choice == '5':
        find_highest_scorer_in_each_room()
    elif choice == '6':
        count_students_passing_all_subjects()
    elif choice == '7':
        break
    else:
        print("เมนูไม่ถูกต้อง กรุณาลองอีกครั้ง")
```
